```python
from browser_env import ScriptBrowserEnv, create_id_based_action
import random

def main():
    env = ScriptBrowserEnv(
        headless=False,  # set True later if you don't want the window
        observation_type="accessibility_tree",
        current_viewport_only=True,
        viewport_size={"width": 1280, "height": 720},
    )

    # Use one of WebArena's example config files
    config_file = r"external\webarena\config_files\examples\test.raw.json"  # path is relative to the webarena repo

    # IMPORTANT: use the correct path if your script is outside the webarena folder
    # e.g. config_file = "external/webarena/config_files/0.json"

    obs, info = env.reset(options={"config_file": config_file})
    print("Initial URL:", info.get("url"))
    print("Initial obs keys:", obs.keys())

    # Take one random action just to see it move
    random_id = random.randint(0, 1000)
    action = create_id_based_action(f"click [{random_id}]")
    obs, reward, terminated, truncated, info = env.step(action)
    print("Step done. Reward:", reward, "Terminated:", terminated, "Truncated:", truncated)

if __name__ == "__main__":
    main()
```
